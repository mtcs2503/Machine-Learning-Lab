```python
import numpy as np
import pandas as pd

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def compute_cost(X, y, theta):
    m = len(y)


    h = sigmoid(X @ theta)


    h = np.clip(h, 1e-15, 1 - 1e-15)

    cost = (-1/m) * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))

    return cost

def compute_gradient(X, y, theta):
    m = len(y)


    h = sigmoid(X @ theta)
    error = h - y
    gradient = (1/m) * (X.T @ error)

    return gradient

def load_and_preprocess_data(file_path):
    print("Loading and preprocessing data...")
    df = pd.read_csv(file_path)
    df.dropna(inplace=True)
    y = df['TenYearCHD'].values.reshape(-1, 1)
    X = df.drop('TenYearCHD', axis=1)
    X = (X - X.mean()) / X.std()
    X = np.hstack((np.ones((X.shape[0], 1)), X))

    print("Data preprocessing complete.")
    print(f"Shape of feature matrix X: {X.shape}")
    print(f"Shape of target vector y: {y.shape}\n")

    return X, y

if __name__ == "__main__":
    file_name = '/content/framingham.csv'

    X, y = load_and_preprocess_data(file_name)

    num_features = X.shape[1]
    theta_initial = np.zeros((num_features, 1))

    initial_cost = compute_cost(X, y, theta_initial)
    initial_gradient = compute_gradient(X, y, theta_initial)

    print("--- Initial values (before any training) ---")
    print(f"Initial Cost (Loss Function): {initial_cost:.4f}")

    print("\nInitial Gradient:")
    print(initial_gradient)


    alpha = 0.01
    theta_updated = theta_initial - alpha * initial_gradient

    updated_cost = compute_cost(X, y, theta_updated)
    updated_gradient = compute_gradient(X, y, theta_updated)

    print("\n--- After one gradient descent step ---")
    print(f"Updated Cost (Loss Function): {updated_cost:.4f}")
    print("\nUpdated Gradient:")
    print(updated_gradient)
```

```python
import pandas as pd
from snorkel.labeling import labeling_function, LFApplier, LabelModel

# Step 1: Load the dataset

df = pd.read_csv('smalldt.csv')

# Separate the labeled and unlabeled data based on the provided example.
# In a real-world scenario, you would have a much larger unlabeled set.
labeled_data = df[df['label'].isin([0, 1, 3])]
unlabeled_data = df[df['label'].isna()]

# Step 2: Define the Labeling Functions (LFs)
# Weak supervision relies on human-defined heuristics. These are examples.
# You would need to create your own based on your domain knowledge.

@labeling_function()
def lf_positive_keywords(x):
    positive_keywords = ['happy', 'glad', 'strong', 'good', 'reassured']
    if any(word in x.text.lower() for word in positive_keywords):
        return 1  # Positive
    return -1

@labeling_function()
def lf_negative_keywords(x):
    negative_keywords = ['humiliated', 'hopeless', 'greedy', 'grouchy', 'pathetic', 'burdened', 'dissatisfied']
    if any(word in x.text.lower() for word in negative_keywords):
        return 0  # Negative
    return -1

@labeling_function()
def lf_negative_words_with_not(x):
    if 'not feel' in x.text.lower() and 'reassured' in x.text.lower():
        return 1  # Positive
    return -1

# You would define many more LFs to cover a wide range of cases...
# For this example, we'll stick to these three.
lfs = [lf_positive_keywords, lf_negative_keywords, lf_negative_words_with_not]

# Step 3: Apply the LFs to the unlabeled data
applier = LFApplier(lfs=lfs)
L_unlabeled = applier.apply(df=unlabeled_data)

# Step 4: Train the Label Model
# The Label Model learns the accuracy and correlations of the LFs to
# produce a single, more reliable label for each data point.
label_model = LabelModel(cardinality=2, verbose=True) # Assuming 2 classes: 0 and 1
label_model.fit(L_train=L_unlabeled, n_epochs=500, log_freq=100, seed=123)

# Step 5: Get the denoised labels for the unlabeled data
unlabeled_data['weak_label'] = label_model.predict(L_unlabeled)

# Print a sample of the results
print("Unlabeled data with weak labels:")
print(unlabeled_data[['text', 'weak_label']].head())
```

```python
from snorkel.labeling import labeling_function
import pandas as pd

# Define our labels. Snorkel uses -1 to represent abstention.
# You can use any integer for your classes.
ABSTAIN = -1
NEGATIVE = 0
POSITIVE = 1
df = pd.read_csv('/content/smalldt.csv')

# Separate the labeled and unlabeled data based on the provided example.
# In a real-world scenario, you would have a much larger unlabeled set.
labeled_data = df[df['label'].isin([0, 1, 3])]
unlabeled_data = df[df['label'].isna()]

@labeling_function()
def lf_negative_keywords(x):
    """Labels a text as negative if it contains certain keywords."""
    negative_words = ['pathetic', 'humiliated', 'horrible', 'greedy', 'hopeless']
    if any(word in x.text.lower() for word in negative_words):
        return NEGATIVE
    return ABSTAIN

@labeling_function()
def lf_positive_keywords(x):
    """Labels a text as positive if it contains positive keywords."""
    positive_words = ['happy', 'fantastic', 'glad', 'reassured', 'energetic']
    if any(word in x.text.lower() for word in positive_words):
        return POSITIVE
    return ABSTAIN

@labeling_function()
def lf_feeling_bad(x):
    """A rule to capture a feeling of sadness or burden."""
    if 'feel' in x.text.lower() and ('burdened' in x.text.lower() or 'guilty' in x.text.lower()):
        return NEGATIVE
    return ABSTAIN

print("Unlabeled data with weak labels:")
print(unlabeled_data[['text', 'weak_label']].head())
```
